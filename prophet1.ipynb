```python
import pandas as pd
from prophet import Prophet
%matplotlib inline
```

```python
df = pd.read_csv('20210826.csv')
```

```python
#모델 생성(유연성 조정)
m = Prophet(changepoint_prior_scale=0.2)
m.fit(df)
```

```python
#미래 데이터프레임 생성
future = m.make_future_dataframe(periods=10)
```

```python
#예측, 출력
forecast = m.predict(future)
forecast.tail(10)
```

```python
#시각화
fig1 = m.plot(forecast)
```
